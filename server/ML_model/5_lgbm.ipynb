```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
```

```python
tourlist=pd.read_excel('dataset_clean.xlsx')
tourlist.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
tourlist['관광지명'] = encoder.fit_transform(tourlist['관광지명'])
tourlist['weather'] = encoder.fit_transform(tourlist['weather'])
tourlist['month'] = encoder.fit_transform(tourlist['month'])
tourlist['weekday'] = encoder.fit_transform(tourlist['weekday'])
```

```python
train = tourlist[['관광지명','avgTemp','highTemp','lowTemp','weather','rainProb','month','weekday']]
target = tourlist['검색건수']
train_input,test_input,train_target,test_target = train_test_split(
    train,target, test_size = 0.3)
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor()
scores = cross_validate(model, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']),np.mean(scores['test_score']))
```

```python
model.fit(train_input,train_target)
print(model.feature_importances_)
```
