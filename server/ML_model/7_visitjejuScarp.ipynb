```python
def fmake_file():
  output_file_name = 'visitjeju_info.csv'
  output_file = open(output_file_name, "w", encoding="utf-8")
  output_file.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(
        'title',
        'contentsid',
        'latitude',
        'longitude',
        'address',
        'roadaddress',
        'tag',
        'alltag',
        'introduction',
        'sbst',
        'imgpath',
        'thumbnailpath',))
  output_file.close()
  return output_file_name
```
