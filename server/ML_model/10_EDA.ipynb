```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('dataset_clean.xlsx')
```

```python
df.info()
```

```python
total_by_month = df[df['datetime'].year==2021]['month']
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(total_by_month.values, total_by_month.index)
ax.set_title('Total Search by Month', fontweight = 'bold', fontsize=16)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
total_by_weather = df['weather']
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(total_by_weather.values, total_by_weather.index)
ax.set_title('Total Search by weather', fontweight = 'bold', fontsize=16)
ax.set_xticks(1,3,4,5,6) #1~6 xlabel입력
# ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'])
sns.despine(ax=ax, top=True, bottom=False, left=True, right=True)
```

```python
total_by_area = df['area_code']
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(total_by_area.values, total_by_area.index)
ax.set_title('Total Search by area', fontweight = 'bold', fontsize=16)
ax.set_xticks(range(0,12))
ax.set_xticklabels(['남원읍', '대정읍', '성산읍', '안덕면', '표선면','서귀포시', '구좌읍', '애월읍', '조천읍', '한경면', '한림읍', '제주시'])
sns.despine(ax=ax, top=True, bottom=False, left=True, right=True)
```
